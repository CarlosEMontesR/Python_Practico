```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#Descargar  los datos
df = yf.download('MSFT', 
                start='1988-01-01', 
                end='2020-12-31',
                auto_adjust=False,
                progress=False)
df = df.loc[:,['Adj Close']]
df.rename(columns={'Adj Close':'adj_close'}, inplace=True)
```

```python
import matplotlib.pylab as plt
import warnings

```

```python
#Calcular los rendimientos simples y logaritmicos utilizando los precios de cierre ajustados
df['simple_rtn'] = df.adj_close.pct_change()
df['log_rtn'] = np.log(df.adj_close/df.adj_close.shift(1))

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
fig, ax = plt.subplots(3,1, figsize=(12,10), sharex=True)
#Agregar precios
df.adj_close.plot(ax=ax[0])
ax[0].set(title = 'MSFT time series',
        ylabel='Stock Price($)')
#Agregar rendimientos simples
df.simple_rtn.plot(ax=ax[1])
ax[1].set(ylabel = 'Simple returns (%)')
#Agregar rendimientos simples
df.simple_rtn.plot(ax=ax[1])
ax[1].set(ylabel = 'Simple returns (%)')
#Agregar rendimientos logaritmicos y graficar
df.log_rtn.plot(ax=ax[2])
ax[2].set(xlabel='Date',
        ylabel='Log returns(%)')

ax[2].tick_params(axis='x',
                which='major',
                labelsize=12)

#plt.tigth_layout()
#plt.savefig()
plt.show()
```
