```python
#Descargamos los datos y solo nos quedamos con el precio ajustado
df = yf.download('AAPL',
                start='2020-01-01',
                end='2024-01-01',
                progress=False)

df = df.loc[:, ['Adj Close']]
df.rename(columns={'Adj Close':'adj_close'}, inplace=True)
```

```python
#Rendimientos simples y log utilizando los precios de cierre ajustados
df['simple_rtn'] = df.adj_close.pct_change()
df['log_rtn'] = np.log(df.adj_close / df.adj_close.shift(1))
df
```
